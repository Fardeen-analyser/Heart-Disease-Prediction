### <span style='color:red'>  1.2.1 Distribution of age

<span style='color:blue'>Showing the count value of age of patient in the dataset, and also see which age appears most. This helps us to understand, the patients are generally younger or older.

```python
#sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})

plt.figure(figsize=(18,7),dpi=100, facecolor='lightgreen')
ax = sns.countplot(data=df,x="age",order=df["age"].value_counts(ascending=False).index,palette="mako")

for bar in ax.containers:
    ax.bar_label(bar, color='brown')
    ax.set_facecolor('lightblue')
plt.title("Patient's age count ",fontsize=20, color='purple')
plt.xlabel("age",fontsize=15, color='indigo')
plt.ylabel("count",fontsize=15, color='indigo')
plt.xticks(rotation=25, color='blue', fontsize=15)
plt.yticks(rotation=25, color='blue', fontsize=15)
plt.show()
```

```python

data_mode = df["age"].mode()
data_mean = df["age"].mean()
```

<span style='color:blue'>  Creating a distribution plot of the "Age" feature and drawing a line representing the mean.

```python
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'})

sns.displot(data=df,x="age",kind='kde', height=6, aspect=2, alpha=0.4, linewidth=2, hue="sex", fill=True,palette="mako", legend=False)

plt.axvline(x=data_mean, ls="--", linewidth=2, color="r")
plt.title("Age distribution on the basis of gender",fontsize=20, color='purple')
plt.xlabel("age", fontsize=15, color='indigo')
plt.ylabel("Density", fontsize=15, color='indigo')
plt.legend(["Male","Female"])
plt.xticks(rotation=25, color='blue', fontsize=15)
plt.yticks(rotation=25, color='blue', fontsize=15)
plt.show()
```

<span style='color:brown'>  The feature age is negatively skewed as observed from the figure, wich means the mean is less than the mode. Of course this tells us that the data contains more elderly people than younger ones.

## <span style='color:red'>  1.4.2 Recitfying feature types


 <span style='color:blue'> We know that some of these features are purely numerical and others are purely categorical. Hence, we need to set the feature type so that our model takes it apporpriately.

```python
scaled_data = scaled_data.astype({'age':float, 'sex':object, 'cp':object, 'trestbps':float, 'chol':float, 'fbs':object, 'restecg':object,'exang':object, 'oldpeak':object, 'slope':object, 'ca':object, 'thal':object,'thalach':float})
```

## <span style='color:red'>  1.4.3 Feature Selection

 <span style='color:blue'> Getting the best features for our model using Chi-square test and SelectKBest method in Sci-Kit Learn.



```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df.drop('target',axis=1)
y = df['target']
best_features = SelectKBest(score_func=chi2,k=10).fit(X,y)
scores = best_features.scores_
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
emp_dict = {}
for i in range(0,len(columns)):
  print(f"Score of Feature {columns[i]} : {scores[i]:.2f}")
  emp_dict.update({columns[i]:scores[i]})
```

 <span style='color:blue'> Hence, the most important features for model building are:

 <span style='color:brown'> 'age'
'chol',
'exang',
'ca',
'cp',
'oldpeak',
'thalach'
and other usable features would be:
 <span style='color:lightbluepink'> 
'trestbps'
'slope'

```python
new_dict = dict(sorted(emp_dict.items(), key=lambda item: item[1]))
```
